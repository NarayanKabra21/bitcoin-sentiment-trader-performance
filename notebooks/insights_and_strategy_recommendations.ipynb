```python
import pandas as pd

df = pd.read_csv("../data/processed/merged_trader_sentiment.csv")
df.columns = df.columns.str.lower()

if "closedpnl" in df.columns:
    df = df.rename(columns={"closedpnl": "pnl"})
elif "closed_pnl" in df.columns:
    df = df.rename(columns={"closed_pnl": "pnl"})

df["date"] = pd.to_datetime(df["date"])
df.head()
```

```python
df["is_win"] = df["pnl"] > 0
df.groupby("classification")["is_win"].mean()
```

```python
loss_threshold = df["pnl"].quantile(0.05)

df[df["pnl"] <= loss_threshold] \
  .groupby("classification")["pnl"] \
  .count()
loss_threshold = df["pnl"].quantile(0.05)

df[df["pnl"] <= loss_threshold] \
  .groupby("classification")["pnl"] \
  .count()
```

```python
df["risk_flag"] = "Normal"
df.loc[df["classification"] == "Greed", "risk_flag"] = "High Risk"

df[["classification", "risk_flag"]].value_counts()
```
