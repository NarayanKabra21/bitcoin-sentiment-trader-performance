```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/merged_trader_sentiment.csv")
df["date"] = pd.to_datetime(df["date"])

df.head()
```

```python
print(df.columns.tolist())
```

```python
df = df.rename(columns={
    "closed_pnl": "pnl",
    "closedpnl": "pnl",
    "closedpnl": "pnl",
    "closedpnl": "pnl",
    "pnl": "pnl"
})
```

```python
df.groupby("classification")["pnl"].mean()
df.groupby("classification")["pnl"].median()
```

```python
df["is_win"] = df["pnl"] > 0
df.groupby("classification")["is_win"].mean()
```

```python
df[df["pnl"] < 0].groupby("classification")["pnl"].describe()
```

```python
df[df["pnl"] > 0] \
  .groupby("classification")["pnl"] \
  .describe()
```

```python
summary = df.groupby("classification").agg(
    avg_pnl=("pnl", "mean"),
    median_pnl=("pnl", "median"),
    win_rate=("pnl", lambda x: (x > 0).mean()),
    loss_rate=("pnl", lambda x: (x < 0).mean())
)

summary
```
